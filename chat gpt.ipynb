```python
import pandas as pd
import numpy as np
loc = r'C:\Users\ASUS\Desktop\Naveen\SFU\Sem_2\Special topics\Project\datasets\rayyan\MV.csv'

# Load your data
data = pd.read_csv(loc)

# Extract a list of non-missing beer names
non_missing_beer_names = data['article_title'].dropna().unique()

# Seed the random number generator for reproducibility
np.random.seed(42)

# Apply a function to fill missing beer_name values with random choices from the non-missing beer names
data['article_title'] = data['article_title'].apply(lambda x: np.random.choice(non_missing_beer_names) if pd.isnull(x) else x)

# Check if there are any missing values left in the beer_name column
print(data['article_title'].isnull().sum())

# Save the modified dataset back to a CSV if needed
data.to_csv('MV_filled.csv', index=False)
data
```
